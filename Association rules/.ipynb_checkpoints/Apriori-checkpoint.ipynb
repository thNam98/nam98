```python
import numpy as np
import pandas as pd

def kiemtra(thn,tn):
    for i in tn:
        if len(i[0]) == 1:
            if thn > set({i}):
                return True
        else:
            if thn > set(i):
                return True
    return False

def duyet(x,y):
    a = 1
    if (len(x[0]) == 1 and x < y):
        return True
    if (len(x[0]) >1):
        if x[len(x)-1] < y[len(x)-1]:
            for i in range(len(x)-1):
                if (x[i] == y[i]):
                    a = a+1
        if a == len(x):
            return True
    return False

def tao_dict(di):
    dict = {}
    for i in range(len(di)):
        dict.update({i : di[i]})
    return(dict)

def tinhchung(n,min_support, dict):
    g=np.sum(n,axis= 0)
    a = []
    b = []
    d = []
    for i,k in enumerate(g):
        if k/n1.shape[0]>=min_support:
            a.append(dict[i])
            b.append(k)
        else:
            d.append(dict[i])
    return(a, b, d, g)

def tao_matran(a,ds,d,nx):
    e = []
    f = []
    for i in a:
        for j in a:
            if duyet(i,j):
                if len(i[0]) == 1:
                    hn = set([j]).union(set([i]))
                else:
                    hn = set(j).union(set(i))
                if kiemtra(hn,d):
                     pass
                else:
                    e.append(sorted(hn))
                    for n in ds:
                        for m in ds:
                            if i == ds[n] and j == ds[m]:
                                 f.append(nx[:,n] & nx[:,m])
                     
    return (e,f)

min_support = 0.05
min_confidence= 0.4
min_lift = 1.2


# In[2]:


a = np.array(pd.read_excel('TranHoangNam.xlsx').fillna(0))

t = list(a[:,1:])
ghn = {}
for i in t:
    ghn= set(i).union(ghn)
ghn.remove(0)
qhn = []
for n in range(len(t)):
    qh = list(np.zeros(len(ghn), dtype = np.int8))
    for i,j in enumerate(ghn):
        for k in t[n]:
            if j == k:
                qh[i] = 1
    qhn.append(qh)

n1 = np.array(qhn)
my_dict = {}
for i in range(len(ghn)):
    my_dict.update({i : list(ghn)[i]})

a1, b1, d1, g1 = tinhchung(n1,min_support, my_dict)

hn=[]
for i,j in enumerate(a1):
    for n,m in enumerate(b1):
        if i == n:
            print('%r: %r'%(j,m))
            hn.append([j,m])


# In[3]:


ax = a1
ds = my_dict
dx = d1
nx = n1
hn1 = hn

for i in ghn:
    ei,fi = tao_matran(ax,ds,dx,nx)
    ds = tao_dict(ei)
    nx= np.array(fi).T
    
    if fi != []:
        ax, bx, dx, gx = tinhchung(nx, min_support, ds)
        hn2 = []
        for i,j in enumerate(ax):
            for n,m in enumerate(bx):
                if (i == n and m/n1.shape[0] > 0):
                    print('%r: %r'%(j,m))
                    hn2.append([j,m])
        print('')
        if hn1 == hn:
            for i in hn2:
                for j in hn:
                    for k in hn:
                        if (k[0] < j[0] and {k[0]} < set(i[0]) and {j[0]} < set(i[0]) and n1.shape[0]*i[1]/j[1]/k[1]>min_lift):
                            if i[1]/k[1]>= min_confidence:
                                print('%r --> %r: supp = %r, conf = %r, lift = %r'%(k[0], j[0], i[1]/n1.shape[0], i[1]/k[1], n1.shape[0]*i[1]/j[1]/k[1]))
                            if i[1]/j[1]>= min_confidence:
                                print('%r --> %r: supp = %r, conf = %r, lift = %r'%(j[0], k[0], i[1]/n1.shape[0], i[1]/j[1], n1.shape[0]*i[1]/j[1]/k[1]))
        else:
            for i in hn2:
                for j in hn1:
                    for k in hn:
                        if ({k[0]} < set(i[0]) and set(j[0]) < set(i[0]) and not({k[0]} <set(j[0])) and n1.shape[0]*i[1]/j[1]/k[1]>min_lift):
                            if i[1]/k[1]>= min_confidence:
                                print('%r --> %r: supp = %r, conf = %r, lift = %r'%(k[0], j[0], i[1]/n1.shape[0], i[1]/k[1], n1.shape[0]*i[1]/j[1]/k[1]))
                            if i[1]/j[1]>= min_confidence:
                                print('%r --> %r: supp = %r, conf = %r, lift = %r'%(j[0], k[0], i[1]/n1.shape[0], i[1]/j[1], n1.shape[0]*i[1]/j[1]/k[1]))
        hn1 = hn2
        print('--------------------------------------------------------------------------------------------------------------------------')
        print('')
    else:
        print("Chúng ta nên dừng lại ở đây thôi")
        break
```
