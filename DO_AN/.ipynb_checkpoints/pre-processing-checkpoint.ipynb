```python
import fiona
import rasterio
import rasterio.mask
import numpy as np
import matplotlib.pyplot as plt
import os
import csv
import pandas as pd
import pprint

src1 = []
src2 = []
with fiona.open("shp/paddy_bg.shp", "r") as shapefile:
    features = [f["properties"] for f in shapefile]
    for i in range(len(features)):
      if(features[i]['class'] == 'background'):
        src1.append(shapefile[i])
      elif(features[i]['class'] == 'paddy'):
        src2.append(shapefile[i])
    geo_paddy = [geo_paddy["geometry"] for geo_paddy in src2]
    geo_background = [geo_background["geometry"] for geo_background in src1]
with rasterio.open("img/img.tif") as src:
    height = src.height
    width = src.width
    src_transform = src.transform
    out_meta = src.meta.copy()
    out_meta.update({'count':1})
mask_paddy = rasterio.features.geometry_mask(geo_paddy, (height, width), src_transform,invert=True, all_touched=True).astype(np.uint32)
mask_background = rasterio.features.geometry_mask(geo_background, (height, width), src_transform,invert=True, all_touched=True).astype(np.uint32)
```

```python
plt.imshow(mask_background)
```

```python
mask = mask_paddy+2*mask_background
mask = mask.flatten().reshape(-1,1) 
```

```python
mask = mask_paddy+2*mask_background
mask = mask.flatten().reshape(-1,1) 
X = []
image_list = os.listdir('img')
for image in image_list:
    with rasterio.open('img/{}'.format(image)) as src:
        X.append(src.read())
X = np.array(X).reshape(4,608,552)
A = [X[i].flatten() for i in range(4)]
A = np.array(A).T
B=np.hstack((A,mask))
C = [B[i,:] for i in range(mask.shape[0]) if(mask[i] > 0)]
C = np.array(C)

df = pd.DataFrame(C, columns = ['ban1', 'ban2', 'ban3', 'ban4', 'paddy'])
df.to_csv('data.csv')
```
