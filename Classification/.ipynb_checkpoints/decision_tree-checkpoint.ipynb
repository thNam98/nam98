```python
import numpy as np
import pandas as pd
import math

df = np.random.permutation(pd.read_csv("TranHoangNam.csv"))


def entropy(x):
    if x.shape[0] == 0:
        return float(0)
    else:
        n = np.sum(x[:,-1])
        return -(float(n/x.shape[0])*log((n/x.shape[0]))+ (1 -float(n/x.shape[0]))*log(1-(n/x.shape[0])))

def log(x):
    if x== 0:
        return float(0)
    else:
        return math.log(x,2)

class exNode:
    def __init__(self, c):
        self.Class = c
class inNode:
    def __init__(self, split_att, split_value, left, right):
        self.split_att = split_att
        self.split_value = split_value
        self.left = left
        self.right = right
        
        
def tree(X,e=0, l=3):
    if e >= l:
        p1 = np.count_nonzero(X[:, -1]) / X.shape[0]
        if p1 >= 0.5:
            pred_class = 1
        else:
            pred_class = 0
        return exNode(pred_class)
    elif np.sum(X[:, -1] - X[0, -1]) == 0:
        return exNode(X[0, -1])
    else:
        entro = float(1)
        left = []
        right = []
        for i in range(4):
            g = sorted(X, key=lambda x: x[i])
            g = np.array(g)
            for j in range(g.shape[0]):
                if (entropy(g[:j,:])*(g[:j,:].shape[0]) + entropy(g[j:,:])*(g[j:,:].shape[0]))/g.shape[0] < entro:
                    entro = (entropy(g[:j,:])*g[:j,:].shape[0] + entropy(g[j:,:])*g[j:,:].shape[0])/g.shape[0]
                    f = (g[j][i]+ g[j-1][i])/2
                    thn = i
                    left = g[:j,:]
                    right = g[j:,:]
        return inNode(thn, f, tree(left, e+1), tree(right, e+1))
    
def printNode(X, T,indent="|--- "):
    if T.__class__ == exNode:
        print(indent + "class:"+ str(T.Class))
    else:
        Xl = X[X[:, T.split_att] < T.split_value]
        Xr = X[X[:, T.split_att] >= T.split_value]
        print(indent+ str(T.split_att) +"<="+ str(T.split_value))
        if Xl.shape[0] > 0:
            printNode(Xl, T.left , "|   "+ indent)
        print(indent+ str(T.split_att) +">"+ str(T.split_value))
        if Xr.shape[0] > 0:
            printNode(Xr, T.right, "|   "+ indent)
    
def main():
    train = df[:100,:]
    train[:,-1][train[:,-1]==2] = 0
    nam = tree(train)
    hoangnam = printNode(train,nam)
    
if __name__ == "__main__":
    main()
```
