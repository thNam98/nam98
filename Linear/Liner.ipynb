```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df = np.array(pd.read_excel('gianha.xlsx'))
x1 = df[:,1]
x2 = df[:,2]
x1 = (x1- x1.min(axis=0))/(x1.max(axis= 0)- x1.min(axis= 0))
x2 = (x2- x2.min(axis=0))/(x2.max(axis= 0)- x2.min(axis= 0))
x = np.hstack((np.ones((df.shape[0], 1)), x1.reshape(-1,1) ,x2.reshape(-1,1)))
y = df[:,0]

y = (y- y.min(axis=0))/(y.max(axis= 0)- y.min(axis= 0))
S = ((np.linalg.inv(x.T.dot(x))).dot(x.T)).dot(y.T)
print(S)
for i,k in enumerate(S):
    print('+Giá trị trọng số w%r: %r'%(i,k))
print("")
RSS = (y - (x.dot(S.T))).T.dot(y - (x.dot(S.T)))
for i,k in enumerate(y - x.dot(S.T)):
    print('-Sai số epsilon %d là: %r'%(i+1,k))
print('+,Tổng bình phương sai số RSS:',RSS)
print('')
TSS = y.dot(y.T) - x.shape[0]*(np.average(y)**2)
# print('+,Tổng độ lệch bình phương TSS:', TSS)
print('-->Giá trị hệ số xác định R square:',1-RSS/TSS)

if np.sqrt(RSS/x.shape[0]) < np.average(y):
    print('==>Mô hình phù hợp')
else:
    print('==>Mô hình ko phù hợp')
```
